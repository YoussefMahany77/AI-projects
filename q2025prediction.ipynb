```python
# def predict_sales(selected_month):
#     month_num = (pd.to_datetime(selected_month).year - df['Date'].dt.year.min()) * 12 + pd.to_datetime(selected_month).month
#     pred = model.predict(np.array([[month_num]]))[0]
#     print(f'Predicted sales for {selected_month}: {pred:.2f}')
#     return pred
#     predict_button = widgets.Button(description='Predict Sales')
# def on_predict_clicked(b):
#     predict_sales(month_dropdown.value)
#     predict_button.on_click(on_predict_clicked)
#     display(predict_button)
# print(predict_sales(month_dropdown.value))
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingRegressor
import matplotlib.pyplot as plt

class EnhancedHybridModel:
    """
    Enhanced Hybrid Time Series Model combining Gradient Boosting and LSTM
    with clear prediction display functionality.
    """
    
    def __init__(self, lookback=12, lstm_units=64, dropout_rate=0.2,
                 n_estimators=100, learning_rate=0.1,
                 epochs=100, batch_size=32):
        # Model parameters
        self.lookback = lookback
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        
        # Initialize models and scalers
        self.lstm_model = None
        self.gb_model = None
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        
        # Track data properties
        self.last_date = None
        self.freq = None
        self.last_sequence = None
        
    def _create_time_features(self, dates):
        """Create time-based features for gradient boosting"""
        if not isinstance(dates, pd.DatetimeIndex):
            dates = pd.DatetimeIndex(dates)
            
        features = pd.DataFrame({
            'year': dates.year,
            'month': dates.month,
            'quarter': dates.quarter,
            'dayofyear': dates.dayofyear,
            'weekofyear': dates.isocalendar().week,
            'is_month_start': dates.is_month_start.astype(int),
            'is_month_end': dates.is_month_end.astype(int),
            'is_quarter_start': dates.is_quarter_start.astype(int),
            'is_quarter_end': dates.is_quarter_end.astype(int),
            'is_year_start': dates.is_year_start.astype(int),
            'is_year_end': dates.is_year_end.astype(int)
        })
        
        # Add cyclical features
        for col in ['month', 'quarter']:
            max_val = 12 if col == 'month' else 4
            features[f'{col}_sin'] = np.sin(2 * np.pi * features[col] / max_val)
            features[f'{col}_cos'] = np.cos(2 * np.pi * features[col] / max_val)
            
        return features
    
    def _create_sequences(self, data):
        """Create sequence data for LSTM"""
        X, y = [], []
        for i in range(len(data) - self.lookback):
            X.append(data[i:(i + self.lookback), 0])
            y.append(data[i + self.lookback, 0])
        return np.array(X), np.array(y)
    
    def _build_lstm_model(self, input_shape):
        """Build LSTM model architecture"""
        model = Sequential()
        model.add(LSTM(units=self.lstm_units, return_sequences=True, 
                       input_shape=input_shape))
        model.add(Dropout(self.dropout_rate))
        model.add(LSTM(units=self.lstm_units//2))
        model.add(Dropout(self.dropout_rate))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mean_squared_error')
        return model
    
    def fit(self, df, date_col='Date', target_col='Sales', verbose=1):
        """
        Fit both models to the training data
        
        Parameters:
        -----------
        df : pandas.DataFrame
            DataFrame with date and target columns
        date_col : str
            Name of the date column
        target_col : str
            Name of the target column
        verbose : int
            Verbosity level (0=silent, 1=progress, 2=detailed)
        """
        if verbose > 0:
            print("Preparing data for hybrid model training...")
        
        # Store properties for forecasting
        self.last_date = df[date_col].max()
        self.freq = pd.infer_freq(df[date_col])
        if self.freq is None:
            self.freq = 'MS'  # Default to month start
            
        # Create features for GB model
        time_features = self._create_time_features(df[date_col])
        
        # Train GB model
        if verbose > 0:
            print("Training Gradient Boosting model...")
            
        self.gb_model = GradientBoostingRegressor(
            n_estimators=self.n_estimators,
            learning_rate=self.learning_rate,
            max_depth=5,
            random_state=42
        )
        self.gb_model.fit(time_features, df[target_col])
        
        # Get GB predictions
        gb_predictions = self.gb_model.predict(time_features).reshape(-1, 1)
        
        # Calculate residuals for LSTM
        residuals = df[target_col].values.reshape(-1, 1) - gb_predictions
        scaled_residuals = self.scaler.fit_transform(residuals)
        
        # Create sequences for LSTM
        X_lstm, y_lstm = self._create_sequences(scaled_residuals)
        X_lstm = X_lstm.reshape(X_lstm.shape[0], X_lstm.shape[1], 1)
        
        # Store last sequence for forecasting
        self.last_sequence = scaled_residuals[-self.lookback:].reshape(1, self.lookback, 1)
        
        # Train LSTM model
        if verbose > 0:
            print("Training LSTM model...")
            
        self.lstm_model = self._build_lstm_model((X_lstm.shape[1], 1))
        
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            ModelCheckpoint('hybrid_model.h5', save_best_only=True)
        ]
        
        history = self.lstm_model.fit(
            X_lstm, y_lstm,
            epochs=self.epochs,
            batch_size=self.batch_size,
            validation_split=0.2,
            verbose=(1 if verbose > 1 else 0),
            callbacks=callbacks
        )
        
        if verbose > 0:
            print("Model training complete!")
            
        return self
    
    def make_future_dataframe(self, periods):
        """
        Create a dataframe of future dates for forecasting
        
        Parameters:
        -----------
        periods : int
            Number of periods to forecast
            
        Returns:
        --------
        future_df : pandas.DataFrame
            DataFrame with future dates
        """
        if self.last_date is None:
            raise ValueError("Model must be fitted before making predictions")
            
        future_dates = pd.date_range(
            start=self.last_date + pd.Timedelta(days=1),
            periods=periods,
            freq=self.freq
        )
        
        return pd.DataFrame({'ds': future_dates})
    
    def predict(self, future_df=None, periods=None):
        """
        Generate predictions with the hybrid model
        
        Parameters:
        -----------
        future_df : pandas.DataFrame
            DataFrame with future dates
        periods : int
            Number of periods to forecast (used if future_df not provided)
            
        Returns:
        --------
        forecast_df : pandas.DataFrame
            DataFrame with predictions
        """
        if self.lstm_model is None or self.gb_model is None:
            raise ValueError("Model must be fitted before making predictions")
            
        if future_df is None and periods is not None:
            future_df = self.make_future_dataframe(periods)
        elif future_df is None and periods is None:
            raise ValueError("Either future_df or periods must be provided")
        
        # Get GB predictions for trend/seasonality
        future_features = self._create_time_features(future_df['ds'])
        gb_predictions = self.gb_model.predict(future_features).reshape(-1, 1)
        
        # Generate LSTM predictions for residuals
        lstm_forecasts = []
        current_sequence = self.last_sequence.copy()
        
        for _ in range(len(future_df)):
            # Predict next residual
            next_pred = self.lstm_model.predict(current_sequence, verbose=0)
            lstm_forecasts.append(next_pred[0, 0])
            
            # Update sequence for next prediction
            current_sequence = np.append(current_sequence[:, 1:, :], 
                                        [[next_pred[0, 0]]], 
                                        axis=1)
        
        # Convert to array and inverse transform
        lstm_forecasts = np.array(lstm_forecasts).reshape(-1, 1)
        lstm_forecasts = self.scaler.inverse_transform(lstm_forecasts)
        
        # Combine predictions
        final_forecasts = gb_predictions + lstm_forecasts
        
        # Create forecast dataframe
        forecast_df = future_df.copy()
        forecast_df['yhat'] = final_forecasts
        forecast_df['trend'] = gb_predictions
        forecast_df['residual'] = lstm_forecasts
        
        # Add confidence intervals (simplified)
        std_dev = np.std(final_forecasts) * 1.96
        forecast_df['yhat_lower'] = forecast_df['yhat'] - std_dev
        forecast_df['yhat_upper'] = forecast_df['yhat'] + std_dev
        
        return forecast_df
    
    def display_prediction_for_month(self, forecast_df, target_date):
        """
        Display the prediction for a specific month clearly
        
        Parameters:
        -----------
        forecast_df : pandas.DataFrame
            DataFrame with forecasts from predict()
        target_date : str or datetime
            The date to find predictions for
            
        Returns:
        --------
        prediction_info : dict
            Dictionary with prediction details
        """
        # Convert target_date to datetime if it's a string
        if isinstance(target_date, str):
            target_date = pd.to_datetime(target_date)
            
        # Find the forecast for the target date
        selected_forecast = forecast_df[forecast_df['ds'] == target_date]
        
        if selected_forecast.empty:
            print(f"⚠️ No forecast available for {target_date.strftime('%Y-%m')}")
            return None
        
        # Get prediction values
        predicted_value = selected_forecast['yhat'].values[0]
        trend_component = selected_forecast['trend'].values[0]
        residual_component = selected_forecast['residual'].values[0]
        lower_bound = selected_forecast['yhat_lower'].values[0]
        upper_bound = selected_forecast['yhat_upper'].values[0]
        
        # Display prediction
        print("\n" + "="*50)
        print(f"📊 FORECAST FOR {target_date.strftime('%B %Y').upper()}")
        print("="*50)
        print(f"📈 Predicted Sales: {predicted_value:.2f}")
        print(f"⚖️ Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print("\nComponent Breakdown:")
        print(f"🔹 Trend/Seasonal (GB): {trend_component:.2f} ({(trend_component/predicted_value*100):.1f}%)")
        print(f"🔹 Residual (LSTM): {residual_component:.2f} ({(residual_component/predicted_value*100):.1f}%)")
        print("="*50)
        
        # Return prediction info as dictionary
        return {
            'date': target_date,
            'prediction': predicted_value,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'trend': trend_component,
            'residual': residual_component
        }
    
    def plot_forecast(self, forecast_df, historical_df=None, 
                     date_col='Date', target_col='Sales',
                     highlight_date=None):
        """
        Plot the forecast with optional historical data
        
        Parameters:
        -----------
        forecast_df : pandas.DataFrame
            DataFrame with forecasts
        historical_df : pandas.DataFrame
            DataFrame with historical data
        date_col : str
            Name of date column in historical data
        target_col : str
            Name of target column in historical data
        highlight_date : datetime
            Specific date to highlight in the forecast
        """
        plt.figure(figsize=(12, 7))
        
        # Plot historical data if provided
        if historical_df is not None:
            plt.plot(historical_df[date_col], historical_df[target_col],
                    color='#3366CC', linewidth=2, label='Historical Data')
            
            # Add vertical line separating historical and forecast
            last_historical_date = historical_df[date_col].max()
            plt.axvline(x=last_historical_date, color='gray', linestyle='--', alpha=0.7)
        
        # Plot forecast
        plt.plot(forecast_df['ds'], forecast_df['yhat'],
                color='#E74C3C', linewidth=2, label='Forecast')
        
        # Plot confidence interval
        plt.fill_between(forecast_df['ds'],
                        forecast_df['yhat_lower'],
                        forecast_df['yhat_upper'],
                        alpha=0.2, color='#E74C3C', label='95% Confidence Interval')
        
        # Highlight specific date if provided
        if highlight_date is not None:
            highlight_forecast = forecast_df[forecast_df['ds'] == highlight_date]
            if not highlight_forecast.empty:
                plt.scatter(highlight_date, highlight_forecast['yhat'].values[0],
                          color='green', s=100, zorder=5, label=f'Forecast for {highlight_date.strftime("%b %Y")}')
        
        # Add title and labels
        plt.title('Hybrid Model Forecast (GB + LSTM)', fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Sales', fontsize=12)
        plt.legend()
        plt.grid(alpha=0.3)
        plt.tight_layout()
        
        return plt

# Complete implementation code
"""
# Prepare data
df = df.reset_index()  # Assuming df has Date in index

# Initialize and train the hybrid model
model = EnhancedHybridModel(
    lookback=12,
    lstm_units=64,
    dropout_rate=0.2,
    n_estimators=100,
    learning_rate=0.05,
    epochs=50
)
model.fit(df, date_col='Date', target_col='Sales')

# Get selected month from dropdown
future_date = pd.to_datetime(month_dropdown.value)

# Create future dataframe and generate forecast
future = model.make_future_dataframe(periods=36)
forecast = model.predict(future)

# Display prediction for selected month
prediction_info = model.display_prediction_for_month(forecast, future_date)

# Visualize the forecast
model.plot_forecast(forecast, df, highlight_date=future_date)
plt.show()
"""
```

```python
# --- Hybrid Model with RandomForestRegressor and LSTM ---
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

class EnhancedHybridModel:
    """
    Enhanced Hybrid Time Series Model combining Random Forest and LSTM
    with clear prediction display functionality.
    """
    def __init__(self, lookback=12, lstm_units=64, dropout_rate=0.2,
                 n_estimators=100, epochs=100, batch_size=32):
        self.lookback = lookback
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.n_estimators = n_estimators
        self.epochs = epochs
        self.batch_size = batch_size
        self.lstm_model = None
        self.gb_model = None
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        self.last_date = None
        self.freq = None
        self.last_sequence = None

    def _create_time_features(self, dates):
        if not isinstance(dates, pd.DatetimeIndex):
            dates = pd.DatetimeIndex(dates)
        features = pd.DataFrame({
            'year': dates.year,
            'month': dates.month,
            'quarter': dates.quarter,
            'dayofyear': dates.dayofyear,
            'weekofyear': dates.isocalendar().week,
            'is_month_start': dates.is_month_start.astype(int),
            'is_month_end': dates.is_month_end.astype(int),
            'is_quarter_start': dates.is_quarter_start.astype(int),
            'is_quarter_end': dates.is_quarter_end.astype(int),
            'is_year_start': dates.is_year_start.astype(int),
            'is_year_end': dates.is_year_end.astype(int)
        })
        for col in ['month', 'quarter']:
            max_val = 12 if col == 'month' else 4
            features[f'{col}_sin'] = np.sin(2 * np.pi * features[col] / max_val)
            features[f'{col}_cos'] = np.cos(2 * np.pi * features[col] / max_val)
        return features

    def _create_sequences(self, data):
        X, y = [], []
        for i in range(len(data) - self.lookback):
            X.append(data[i:(i + self.lookback), 0])
            y.append(data[i + self.lookback, 0])
        return np.array(X), np.array(y)

    def _build_lstm_model(self, input_shape):
        model = Sequential()
        model.add(LSTM(units=self.lstm_units, return_sequences=True, input_shape=input_shape))
        model.add(Dropout(self.dropout_rate))
        model.add(LSTM(units=self.lstm_units//2))
        model.add(Dropout(self.dropout_rate))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mean_squared_error')
        return model

    def fit(self, df, date_col='Date', target_col='Sales', verbose=1):
        if verbose > 0:
            print("Preparing data for hybrid model training...")
        self.last_date = df[date_col].max()
        self.freq = pd.infer_freq(df[date_col])
        if self.freq is None:
            self.freq = 'MS'
        time_features = self._create_time_features(df[date_col])
        if verbose > 0:
            print("Training Random Forest model...")
        self.gb_model = RandomForestRegressor(
            n_estimators=self.n_estimators,
            max_depth=5,
            random_state=42
        )
        self.gb_model.fit(time_features, df[target_col])
        gb_predictions = self.gb_model.predict(time_features).reshape(-1, 1)
        residuals = df[target_col].values.reshape(-1, 1) - gb_predictions
        scaled_residuals = self.scaler.fit_transform(residuals)
        X_lstm, y_lstm = self._create_sequences(scaled_residuals)
        X_lstm = X_lstm.reshape(X_lstm.shape[0], X_lstm.shape[1], 1)
        self.last_sequence = scaled_residuals[-self.lookback:].reshape(1, self.lookback, 1)
        if verbose > 0:
            print("Training LSTM model...")
        self.lstm_model = self._build_lstm_model((X_lstm.shape[1], 1))
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            ModelCheckpoint('hybrid_model.h5', save_best_only=True)
        ]
        self.lstm_model.fit(
            X_lstm, y_lstm,
            epochs=self.epochs,
            batch_size=self.batch_size,
            validation_split=0.2,
            verbose=0,
            callbacks=callbacks
        )
        if verbose > 0:
            print("Model training complete!")
        return self

    def make_future_dataframe(self, periods):
        if self.last_date is None:
            raise ValueError("Model must be fitted before making predictions")
        future_dates = pd.date_range(
            start=self.last_date + pd.Timedelta(days=1),
            periods=periods,
            freq=self.freq
        )
        return pd.DataFrame({'ds': future_dates})

    def predict(self, future_df=None, periods=None):
        if self.lstm_model is None or self.gb_model is None:
            raise ValueError("Model must be fitted before making predictions")
        if future_df is None and periods is not None:
            future_df = self.make_future_dataframe(periods)
        elif future_df is None and periods is None:
            raise ValueError("Either future_df or periods must be provided")
        future_features = self._create_time_features(future_df['ds'])
        gb_predictions = self.gb_model.predict(future_features).reshape(-1, 1)
        lstm_forecasts = []
        current_sequence = self.last_sequence.copy()
        for _ in range(len(future_df)):
            next_pred = self.lstm_model.predict(current_sequence, verbose=0)
            lstm_forecasts.append(next_pred[0, 0])
            current_sequence = np.append(current_sequence[:, 1:, :], [[next_pred[0, 0]]], axis=1)
        lstm_forecasts = np.array(lstm_forecasts).reshape(-1, 1)
        lstm_forecasts = self.scaler.inverse_transform(lstm_forecasts)
        final_forecasts = gb_predictions + lstm_forecasts
        forecast_df = future_df.copy()
        forecast_df['yhat'] = final_forecasts
        forecast_df['trend'] = gb_predictions
        forecast_df['residual'] = lstm_forecasts
        std_dev = np.std(final_forecasts) * 1.96
        forecast_df['yhat_lower'] = forecast_df['yhat'] - std_dev
        forecast_df['yhat_upper'] = forecast_df['yhat'] + std_dev
        return forecast_df

    def display_prediction_for_month(self, forecast_df, target_date):
        if isinstance(target_date, str):
            target_date = pd.to_datetime(target_date)
        selected_forecast = forecast_df[forecast_df['ds'] == target_date]
        if selected_forecast.empty:
            print(f"⚠️ No forecast available for {target_date.strftime('%Y-%m')}")
            return None
        predicted_value = selected_forecast['yhat'].values[0]
        trend_component = selected_forecast['trend'].values[0]
        residual_component = selected_forecast['residual'].values[0]
        lower_bound = selected_forecast['yhat_lower'].values[0]
        upper_bound = selected_forecast['yhat_upper'].values[0]
        print("\n" + "="*50)
        print(f"📊 FORECAST FOR {target_date.strftime('%B %Y').upper()}")
        print("="*50)
        print(f"📈 Predicted Sales: {predicted_value:.2f}")
        print(f"⚖️ Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print("\nComponent Breakdown:")
        print(f"🔹 Trend/Seasonal (RF): {trend_component:.2f} ({(trend_component/predicted_value*100):.1f}%)")
        print(f"🔹 Residual (LSTM): {residual_component:.2f} ({(residual_component/predicted_value*100):.1f}%)")
        print("="*50)
        return {
            'date': target_date,
            'prediction': predicted_value,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound,
            'trend': trend_component,
            'residual': residual_component
        }

    def plot_forecast(self, forecast_df, historical_df=None, 
                     date_col='Date', target_col='Sales',
                     highlight_date=None):
        plt.figure(figsize=(12, 7))
        if historical_df is not None:
            plt.plot(historical_df[date_col], historical_df[target_col],
                    color='#3366CC', linewidth=2, label='Historical Data')
            last_historical_date = historical_df[date_col].max()
            plt.axvline(x=last_historical_date, color='gray', linestyle='--', alpha=0.7)
        plt.plot(forecast_df['ds'], forecast_df['yhat'],
                color='#E74C3C', linewidth=2, label='Forecast')
        plt.fill_between(forecast_df['ds'],
                        forecast_df['yhat_lower'],
                        forecast_df['yhat_upper'],
                        alpha=0.2, color='#E74C3C', label='95% Confidence Interval')
        if highlight_date is not None:
            highlight_forecast = forecast_df[forecast_df['ds'] == highlight_date]
            if not highlight_forecast.empty:
                plt.scatter(highlight_date, highlight_forecast['yhat'].values[0],
            color='green', s=100, zorder=5, label=f"Forecast for {highlight_date.strftime('%b %Y')}")
                
        plt.title('Hybrid Model Forecast (Random Forest + LSTM)', fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Sales', fontsize=12)
        plt.legend()
        plt.grid(alpha=0.3)
        plt.tight_layout()
        return plt

# --- Example usage (assuming df is already prepared) ---
# model = EnhancedHybridModel(lookback=12, lstm_units=64, dropout_rate=0.2, n_estimators=100, epochs=50)
# model.fit(df, date_col='Date', target_col='Sales')
# future = model.make_future_dataframe(periods=12)
# forecast = model.predict(future)
# model.display_prediction_for_month(forecast, '2025-06-01')
# model.plot_forecast(forecast, df, highlight_date=pd.to_datetime('2025-06-01'))
# plt.show()
```

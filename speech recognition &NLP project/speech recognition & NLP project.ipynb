```python
import pyttsx3
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser
import os
import random
import tkinter as tk

engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

def wishMe():
    hour = int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("good morning!")
    elif hour>=12 and hour<18:
        speak("good afternoon!")
    else:
        speak("good evening!")

    speak("I am  your virtual assistant. Hello sir?")

def listen():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        speak(" I am Listening")
        r.pause_threshold = 1
        audio = r.listen(source)

    try:
        print("Recognizing...")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")
        process_query(query)

    except Exception as e:
        print(e)
        speak("Say that again please...")

def process_query(query):
    query = query.lower()

    if 'wikipedia' in query:
        speak('Searching Wikipedia...')
        query = query.replace("wikipedia", "")
        results = wikipedia.summary(query, sentences=2)
        speak("According to Wikipedia")
        print(results)
        speak(results)
        
    elif 'hello' in query:
        speak("hello, how are you doing")
        
    elif ' I am fine how are you doing ' in query:
        speak("I am fine thank you ")   

    elif 'open youtube' in query:
        speak("opening youtube")
        webbrowser.open("youtube.com")

    elif 'open google' in query:
        speak("opening google")
        webbrowser.open("google.com")

    elif 'play music' in query:
        speak("playing music")
        music_dir = 'F:\\music\\Playlists'
        songs = os.listdir(music_dir)
        print(songs)
        os.startfile(os.path.join(music_dir, songs[0]))

    elif 'play next music' in query:
        speak("playing another music")
        music_dir = 'F:\\music\\Playlists'
        songs = os.listdir(music_dir)
        print(songs)
        os.startfile(os.path.join(music_dir, songs[random.randint(0, len(songs)-1)]))

    elif 'the time' in query:
        strTime = datetime.datetime.now().strftime("%H:%M:%S")
        speak(f"Sir, the time is {strTime}")
    elif 'launch epic games' in query:
        os.startfile("C:\\Program Files (x86)\\Epic Games\\Launcher\\Portal\\Binaries\\Win32\\EpicGamesLauncher.exe")

    elif 'open discord' in query:
        os.startfile("C:\\Users\\yosse\\AppData\\Local\\Discord\\app-1.0.9013\\Discord.exe")
    
    elif 'goodbye' in query:
        speak("Goodbye!")
        exit()

if __name__ == "__main__":
    wishMe()

    # Create the GUI window
    root = tk.Tk()
    root.title("Virtual Assistant")

    listen_button = tk.Button(root, text="Hold to Speak", bg="blue", fg="black",width=100)
    listen_button.bind("<Button-1>", lambda event: listen())
    listen_button.bind("<ButtonRelease-1>", lambda event: listen_button.configure(bg="blue"))

    listen_button.pack(fill="x")

    # Start the main event loop
    root.mainloop()
```
